## Goal:  predict the tree according to the other features, including original categories?

```python
import os
import pandas as pd
data_path = os.path.abspath(
    os.path.join(
        os.path.pardir,
        'data',
        'prods.tar.gz'
    )
)
data = pd.read_csv(data_path)
data.info()
```

```python
def feature_unique_value_ratio(data,series_name):
    return round(100*data.loc[:,series_name].isnull().sum()/len(data),1)

def produce_unique_value_summary(data,ifdropna):
    if ifdropna==True:
        output= data.\
        describe(include="all").\
        loc['unique',:].\
        dropna().\
        apply(lambda x: round(100*x/len(data),1))
    else :
        output = data.\
        describe(include="all").\
        loc['unique',:].\
        apply(lambda x: round(100*x/len(data),1))
    return output

data1 = data.drop(['ean','url','provider','price','./._prods.csv'],axis=1)
unique_value_summary_nadropped = produce_unique_value_summary(data1,True)
unique_value_summary_withnan = produce_unique_value_summary(data1,False)

print(f"With NA dropped, the summary looks like {unique_value_summary_nadropped}")
#print(f"With NA values, the summary looks like {unique_value_summary_withnan}")
print(data1.head())
```
